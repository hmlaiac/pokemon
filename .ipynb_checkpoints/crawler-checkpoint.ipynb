```python
import requests
from bs4 import BeautifulSoup
```

```python
link = "https://wiki.52poke.com/zh-hant/%E5%AE%9D%E5%8F%AF%E6%A2%A6%E5%88%97%E8%A1%A8%EF%BC%88%E6%8C%89%E5%B8%95%E5%BA%95%E4%BA%9A%E5%9B%BE%E9%89%B4%E7%BC%96%E5%8F%B7%EF%BC%89"

response = requests.get(link)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
tables = soup.find_all("table")
# result = soup.find_all("table")
```

```python
trs = []
for table in tables:
#     tbodys.append(table.find_all("tbody"))
    trs.extend(table.find("tbody").find_all("tr"))
```

```python
for tr in trs:
    sels = tr.select("td")
    if sels:
        print(sels[0].text)
```
